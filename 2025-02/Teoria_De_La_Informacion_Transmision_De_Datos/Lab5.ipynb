## Simulación del canal BSC

Este código realiza una simulación práctica de un canal binario simétrico (BSC) para contrastar los resultados teóricos con datos experimentales. Se genera una secuencia de bits equiprobables y se transmiten a través del canal, en el cual cada bit puede invertirse con una probabilidad de error $f$. La salida se obtiene aplicando una operación XOR entre la señal transmitida y el ruido. A partir de los datos generados se calcula la probabilidad de error de bit.

```python
import numpy as np

def simulate_BSC(N=100000, f=0.01):
    # Fuente equiprobable
    X = np.random.randint(0, 2, N)
    # Canal BSC
    noise = np.random.rand(N) < f
    Y = np.bitwise_xor(X, noise.astype(int))
    return X, Y

# Ejemplo con f=0.01
f = 0.01
N = 100000
X, Y = simulate_BSC(N, f)

# Proporción de bits correctos
prop_error_bit = np.mean(X != Y)

print(f"Probabilidad de error de bit: {prop_error_bit:.4f}")
```
