#### **Árboles de derivación**

Los árboles de derivación son representaciones visuales utilizadas para mostrar cómo una gramática genera una palabra. Nos ayudan a:

✅ Entender cómo se aplican las reglas de producción de una gramática.

✅ Visualizar la estructura de las palabras generadas.

✅ Detectar ambigüedades en una gramática.

##### **Ejemplo 1: Generación de una palabra con una gramática**

**Descripción del problema:**

Queremos simular en Python cómo se generan palabras a partir de una gramática, explorando todas las posibles derivaciones y encontrando la correcta.

**Objetivo:**
- Implementar una derivación paso a paso de una palabra usando una gramática.
- Asegurar que todas las reglas de producción se apliquen correctamente.

**Gramática dada**
- $G = (\{S, A\}, \{a, b\}, S, \{S \to aA, A \to b\})$
- Construir la derivación para la palabra "$ab$".

**Entradas:**
1. Un símbolo inicial $S$.
2. Reglas de producción de la gramática.
3. Palabra objetivo.

**Salidas:**
1. Secuencia de pasos que muestran la derivación.

**Ejemplo de salida esperada:**

```yaml
Paso 1: S  
Paso 2: aA  
Paso 3: ab  
```

````python
from collections import deque

def derivacion(simbolo_inicial, reglas, palabra_objetivo):
    # Fila para explorar todas las posibilidades
    fila = deque([(simbolo_inicial, [simbolo_inicial])])

    while fila:
        actual, pasos = fila.popleft()  # Tomamos el primer elemento de la fila

        # Si hemos alcanzado la palabra objetivo, retornamos los pasos
        if actual == palabra_objetivo:
            return pasos

        for i, simbolo in enumerate(actual):
            if simbolo in reglas:  # Si encontramos un no terminal, aplicamos sus reglas de producción
                for produccion in reglas[simbolo]:
                    nueva_cadena = actual[:i] + produccion + actual[i+1:]  # Sustituimos el símbolo

                    # Garantizamos que la nueva cadena no supere la longitud de la palabra esperada
                    if len(nueva_cadena) <= len(palabra_objetivo):
                        fila.append((nueva_cadena, pasos + [nueva_cadena]))  # Añadimos a la fila para continuar la exploración

    print("No se puede derivar la palabra objetivo con la gramática dada.")
    return []

# Definición de la gramática
gramatica = {
    "S": ["aA"],  # S → aA
    "A": ["b"]    # A → b
}

# Generando la derivación para "ab"
derivacion_resultado = derivacion("S", gramatica, "ab")

# Mostrando los pasos de la derivación
if derivacion_resultado:
    for i, paso in enumerate(derivacion_resultado):
        print(f"Paso {i+1}: {paso}")
````

##### **Ejercicio 2 - Desafío: Mejorando la derivación de palabras**

**Descripción del problema:**

El código de derivación implementado en el Ejercicio 1 genera palabras correctamente en muchos casos. Sin embargo, algunas gramáticas pueden presentar desafíos que impiden su correcto funcionamiento.

Tu tarea es analizar el código, identificar posibles limitaciones y proponer soluciones.

**Objetivo:**

- Identificar casos en los que el código no produce la palabra esperada.

````python
from collections import deque

def derivacion(simbolo_inicial, reglas, palabra_objetivo):
    # Fila para explorar todas las posibilidades
    fila = deque([(simbolo_inicial, [simbolo_inicial])])
    # Conjunto para evitar procesar cadenas repetidas
    procesadas = set()

    while fila:
        actual, pasos = fila.popleft()  # Tomamos el primer elemento de la fila

        # Si hemos alcanzado la palabra objetivo, retornamos los pasos
        if actual == palabra_objetivo:
            return pasos

        # Si ya procesamos esta cadena, la saltamos
        if actual in procesadas:
            continue
        procesadas.add(actual)  # Marcamos la cadena como procesada

        # Recorremos la cadena actual para buscar símbolos no terminales
        for i, simbolo in enumerate(actual):
            if simbolo in reglas:  # Si encontramos un no terminal, aplicamos sus reglas de producción
                for produccion in reglas[simbolo]:
                    # Sustituimos el símbolo no terminal por su producción
                    nueva_cadena = actual[:i] + produccion + actual[i+1:]

                    # Verificamos si la nueva cadena puede generar la palabra objetivo
                    # Eliminamos los símbolos no terminales para comparar con la palabra objetivo
                    cadena_sin_no_terminales = nueva_cadena
                    for no_terminal in reglas:
                        cadena_sin_no_terminales = cadena_sin_no_terminales.replace(no_terminal, "")

                    # Si la cadena sin no terminales coincide con el inicio de la palabra objetivo, continuamos
                    if palabra_objetivo.startswith(cadena_sin_no_terminales):
                        fila.append((nueva_cadena, pasos + [nueva_cadena]))  # Añadimos a la fila para continuar la exploración

    print("No se puede derivar la palabra objetivo con la gramática dada.")
    return []

# Definición de la gramática
gramatica = {
    "S": ["aAb", "aaAb"],  # S → aAb | aaAb
    "A": ["aabbA", "ba"]    # A → aabbA | ba
}

# Generando la derivación para "abab"
derivacion_resultado = derivacion("S", gramatica, "abab")

# Mostrando los pasos de la derivación
if derivacion_resultado:
    for i, paso in enumerate(derivacion_resultado):
        print(f"Paso {i+1}: {paso}")
````

#### **Derivación izquierda y derecha**

Existen dos formas de aplicar las reglas de producción de una gramática:

✅ Derivación por la izquierda: Siempre se sustituye el símbolo no terminal más a la izquierda.

✅ Derivación por la derecha: Siempre se sustituye el símbolo no terminal más a la derecha.

El tipo de derivación afecta la construcción de los árboles de derivación.

##### **Ejemplo 2: Implementación de derivación izquierda y derecha**

**Descripción del problema:**

Queremos construir una palabra a partir de una gramática aplicando dos estrategias diferentes: derivación izquierda y derivación derecha.

**Objetivo:**
- Implementar la derivación de una palabra de forma determinada por el orden de sustitución.
- Comparar los resultados obtenidos con derivación izquierda y derecha.

**Gramática dada:**
- $G = (\{S, A\}, \{a, b\}, S, \{S \to aAS\, |\, a, A \to Ab\,|\,b\})$
- Generar la palabra "$aba$".

**Entradas**
1. Un símbolo inicial S.
2. Reglas de producción de la gramática.
3. Estrategia de derivación (izquierda o derecha).

**Salidas**
1. Secuencia de pasos en la derivación.

**Ejemplo de salida esperada:**
- Derivación por la izquierda:
```yaml
Paso 1: S
Paso 2: aAS
Paso 3: abS
Paso 4: aba
```
- Derivación por la derecha:
```yaml
Paso 1: S
Paso 2: aAS
Paso 3: aAa
Paso 4: aba   
```

````python
from collections import deque

def derivacion(simbolo_inicial, reglas, palabra_objetivo, estrategia="izquierda"):
    # Cola para explorar todas las posibilidades
    fila = deque([(simbolo_inicial, [simbolo_inicial])])

    while fila:
        actual, pasos = fila.popleft()  # Tomamos el primer elemento de la fila

        # Si alcanzamos la palabra objetivo, retornamos los pasos
        if actual == palabra_objetivo:
            return pasos

        # Decidir el orden de sustitución según la estrategia
        indices = range(len(actual)) if estrategia == "izquierda" else reversed(range(len(actual)))

        for i in indices:
            simbolo = actual[i]
            if simbolo in reglas:  # Si encontramos un no terminal, aplicamos sus reglas de producción
                for produccion in reglas[simbolo]:
                    nueva_cadena = actual[:i] + produccion + actual[i+1:]  # Sustituimos el símbolo

                    # Aseguramos que la nueva cadena no sea más larga que la palabra objetivo
                    if len(nueva_cadena) <= len(palabra_objetivo):
                        fila.append((nueva_cadena, pasos + [nueva_cadena]))  # Añadimos a la cola para continuar la exploración
                break  # Aplicamos solo una sustitución por iteración

    print("No se puede derivar la palabra objetivo con la gramática dada.")
    return []

# Definición de la gramática
gramatica = {
    "S": ["aAS", "a"],  # S → aA
    "A": ["Ab", "b"]  # A → Ab | b
}

# Generando la derivación para "aba" con ambas estrategias
print("Derivación por la izquierda:")
derivacion_izquierda = derivacion("S", gramatica, "aba", "izquierda")
for i, paso in enumerate(derivacion_izquierda):
    print(f"Paso {i+1}: {paso}")

print("\nDerivación por la derecha:")
derivacion_derecha = derivacion("S", gramatica, "aba", "derecha")
for i, paso in enumerate(derivacion_derecha):
    print(f"Paso {i+1}: {paso}")
````

````python
from collections import deque

def derivacion(simbolo_inicial, reglas, palabra_objetivo):
    visitados = set()
    fila = deque([(simbolo_inicial, [simbolo_inicial])])
    visitados.add(simbolo_inicial)
    target_len = len(palabra_objetivo)

    while fila:
        actual, pasos = fila.popleft()

        if actual == palabra_objetivo:
            return pasos

        terminales_actual = [c for c in actual if c not in reglas]

        # Verificar coincidencia de terminales con el objetivo
        if len(terminales_actual) > target_len:
            continue

        if terminales_actual != list(palabra_objetivo[:len(terminales_actual)]):
            continue

        for i, simbolo in enumerate(actual):
            if simbolo in reglas:
                for produccion in reglas[simbolo]:
                    nueva_cadena = actual[:i] + produccion + actual[i+1:]

                    # Verificar límite de longitud de terminales
                    terminales_nuevos = [c for c in nueva_cadena if c not in reglas]
                    if len(terminales_nuevos) > target_len:
                        continue

                    # Verificar coincidencia de prefijo
                    if terminales_nuevos != list(palabra_objetivo[:len(terminales_nuevos)]):
                        continue

                    # Evitar ciclos y reprocesamiento
                    if nueva_cadena not in visitados:
                        visitados.add(nueva_cadena)
                        fila.append((nueva_cadena, pasos + [nueva_cadena]))

    print("No se puede derivar la palabra objetivo con la gramática dada.")
    return []

# Ejemplo de uso
gramatica = {
    "S": ["aAb", "aaAb"],
    "A": ["aabbA", "ba"]
}

derivacion_resultado = derivacion("S", gramatica, "abab")

if derivacion_resultado:
    for i, paso in enumerate(derivacion_resultado):
        print(f"Paso {i+1}: {paso}")

####################################################################################################################################################
````

````python
from collections import deque

def derivaciones_completas(simbolo_inicial, reglas, palabra_objetivo):
    # Estructura para guardar las derivaciones encontradas
    derivaciones_encontradas = []

    # Cada elemento en la cola será (cadena, pasos, izquierda_derivada, derecha_derivada)
    fila = deque([(simbolo_inicial, [], False, False)])

    # Diccionario para registrar visitados con sus estados de derivación
    visitados = {}

    while fila:
        actual, pasos, izq_deriv, der_deriv = fila.popleft()

        # Si encontramos la palabra objetivo, guardamos la derivación
        if actual == palabra_objetivo:
            derivaciones_encontradas.append(pasos)
            continue

        # Verificar si ya hemos procesado esta cadena con estos flags de derivación
        clave_visita = (actual, izq_deriv, der_deriv)
        if clave_visita in visitados:
            continue
        visitados[clave_visita] = True

        # Verificar longitud de terminales
        terminales = [c for c in actual if c not in reglas]
        if len(terminales) > len(palabra_objetivo):
            continue

        # Buscar no terminales para derivar
        no_terminales = [(i, c) for i, c in enumerate(actual) if c in reglas]

        # Derivación por izquierda (primer no terminal encontrado)
        if not izq_deriv and no_terminales:
            izquierda_i, izquierda_nt = no_terminales[0]
            for produccion in reglas[izquierda_nt]:
                nueva_cadena = actual[:izquierda_i] + produccion + actual[izquierda_i+1:]
                nuevos_pasos = pasos + [(nueva_cadena, 'Izquierda', izquierda_nt)]
                fila.append((nueva_cadena, nuevos_pasos, True, der_deriv))

        # Derivación por derecha (último no terminal encontrado)
        if not der_deriv and no_terminales:
            derecha_i, derecha_nt = no_terminales[-1]
            for produccion in reglas[derecha_nt]:
                nueva_cadena = actual[:derecha_i] + produccion + actual[derecha_i+1:]
                nuevos_pasos = pasos + [(nueva_cadena, 'Derecha', derecha_nt)]
                fila.append((nueva_cadena, nuevos_pasos, izq_deriv, True))

    return derivaciones_encontradas

# Ejemplo de uso
gramatica = {
    "S": ["aAb", "aaAb"],
    "A": ["aabbA", "ba"]
}

palabra_objetivo = "abab"
resultados = derivaciones_completas("S", gramatica, palabra_objetivo)

if resultados:
    print(f"Se encontraron {len(resultados)} derivaciones posibles para '{palabra_objetivo}':\n")
    for i, derivacion in enumerate(resultados, 1):
        print(f"Derivación {i}:")
        print("Paso 0: S")
        for j, (cadena, direccion, no_terminal) in enumerate(derivacion, 1):
            print(f"Paso {j}: {cadena} (Derivando {no_terminal} por {direccion})")
        print()
else:
    print(f"No se encontraron derivaciones para '{palabra_objetivo}'")
````

from collections import deque

def integracion(palabra_objetivo, reglas, simbolo_inicial):
    # Invertir las reglas para integración
    reglas_invertidas = {}
    for simbolo, producciones in reglas.items():
        for produccion in producciones:
            for i in range(len(produccion)):
                if produccion[i] == simbolo:
                    prefijo = produccion[:i]
                    sufijo = produccion[i+1:]
                    if (prefijo + sufijo) not in reglas_invertidas:
                        reglas_invertidas[prefijo + sufijo] = []
                    reglas_invertidas[prefijo + sufijo].append(simbolo)

    visitados = set()
    fila = deque([(palabra_objetivo, [palabra_objetivo])])
    visitados.add(palabra_objetivo)

    while fila:
        actual, pasos = fila.popleft()

        if actual == simbolo_inicial:
            return pasos

        # Buscar posibles integraciones
        for i in range(len(actual)):
            for j in range(i+1, len(actual)+1):
                subcadena = actual[i:j]
                if subcadena in reglas_invertidas:
                    nueva_cadena = actual[:i] + reglas_invertidas[subcadena][0] + actual[j:]

                    # Evitar ciclos y reprocesamiento
                    if nueva_cadena not in visitados:
                        visitados.add(nueva_cadena)
                        fila.append((nueva_cadena, pasos + [nueva_cadena]))

    print("No se puede integrar hacia el símbolo inicial con la gramática dada.")
    return []
