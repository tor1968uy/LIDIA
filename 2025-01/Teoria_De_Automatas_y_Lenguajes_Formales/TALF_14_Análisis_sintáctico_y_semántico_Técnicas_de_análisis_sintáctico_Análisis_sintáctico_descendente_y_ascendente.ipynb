#### **Análisis sintáctico descendente y ascendente — Comparación en Python**

En esta práctica, vamos a comparar dos formas de realizar el análisis sintáctico:
- **Análisis descendente (top-down)**: parte del símbolo inicial y aplica reglas de producción hasta alcanzar la cadena de entrada.
- **Análisis ascendente (bottom-up)**: parte de la cadena de entrada y aplica reducciones hasta reconstruir la derivación.

Ambos métodos buscan verificar si una palabra puede ser generada por una gramática dada. A veces uno de ellos es más adecuado o eficiente dependiendo de la estructura de la gramática.

Vamos a estudiar un ejemplo concreto con código y luego resolver un ejercicio comparativo.

```python
from collections import deque

# Gramática compatible con derivación por la izquierda
gramatica_desc = {
    "S": [["a", "S", "b"], ["a", "b"]]

                    if len(nueva) <= len(palabra):
                        fila.append((nueva, pasos + [nueva]))
                break
    return []

# Prueba
```

**Análisis de compatibilidad de las gramáticas con los dos métodos de parsing**

**¿Por qué la gramática del análisis descendente (S → aSb ∣ ab) es difícil de usar con el parser ascendente provisto?**

      La gramática S → aSb | ab es recursiva (a la izquierda y a la derecha).
      Esta forma es natural para análisis top-down, porque comienza desde el
      símbolo inicial y va expandiendo, anticipando la estructura esperada (por
      ejemplo, empieza con `a`, y luego busca emparejar cada `a` con un `b` correspondiente).

      En cambio, la logica del parser bottom-up trabaja construyendo desde los
      símbolos terminales hacia la raíz. En este caso, los `a` y `b` se mezclan
      de forma anidada (`a a b b`), por lo que el parser ascendente tendría
      dificultades en decidir "cuándo" reducir "ab" a "S", ya que podría ser
      parte de una derivación más grande como "aSb".
      
      Esta ambigüedad en los puntos de reducción hace que esta gramática sea
      "difícil" de parsear con un parser ascendente.

**¿Por qué la gramática del análisis ascendente (E → E+a ∣ a) funciona bien con esa estrategia?**

      Esta gramática es recursiva "a la izquierda", lo cual es una
      característica típica (y favorable) para los analizadores bottom-up. El
      parser puede ir apilando símbolos ("a", "+", "a", etc.), y luego ejecutar
      reducciones:
        Cada vez que detecta "E + a", puede reducir eso a un nuevo "E".
        Esta forma se adapta bien a la logica que se aplica al algoritmo del parser ascendente.

**¿Qué característica estructural tiene esta segunda gramática?**

      La gramática E → E + a ∣ a es una gramática recursiva por la izquierda.
      La gramatica del ejemplo presenta ademas:
          * Una base clara ("E → a")
          * No es ambigua.

¿Qué ocurre con la derivación descendente?

El algoritmo genera derivaciones válidas porque implementa una estrategia de búsqueda por ancho controlado (no es una derivación descendente recursiva pura).
Gracias al límite de longitud impuesto (if len(nueva) <= len(palabra)), explora solo derivaciones posibles para llegar a la palabra final, evitando ciclos infinitos.

¿Dónde se queda "atascado" el análisis?

En un parser descendente recursivo típico, la regla E → E + a sí causaria una recursión infinita porque el llamado recursivo a E ocurre "antes" de consumir un simbolo terminal.
En el algoritmo propuesto esto no ocurre porque es iterativo y posee una "cola" para elegir que procesar y ademas para evitar recursion infinita posee la condicion que controla el largo de la palabra (condición de poda).

¿Cómo influye la recursividad por la izquierda?

La recursividad por la izquierda es problemática en parsers recursivos "puros", dado que este usa criterios diferentes no "sufre" de esos problemas ya que:

1) Expande solo un símbolo a la vez.

2) Controla la profundidad de la derivación mediante la sentencia len(nueva) <= len(palabra).

3) Evita la expansión indefinida con el uso de un deque y que las derivaciones  mayores a la palabra deseada son ignoradas.

````python
from collections import deque

# Gramática compatible con derivación por la izquierda
gramatica = {
    "E": [["E", "+", "a"], ["a"]]
}

def derivar(simbolo_inicial, reglas, palabra, estrategia="izquierda"):
    fila = deque([([simbolo_inicial], [simbolo_inicial])])
    while fila:
        actual, pasos = fila.popleft()
        if actual == list(palabra):
            return pasos
        indices = range(len(actual)) if estrategia == "izquierda" else reversed(range(len(actual)))
        for i in indices:
            simbolo = actual[i]
            if simbolo in reglas:
                for produccion in reglas[simbolo]:
                    nueva = actual[:i] + produccion + actual[i+1:]
                    if len(nueva) <= len(palabra):
                        fila.append((nueva, pasos + [nueva]))
                break
    return []

# Prueba
palabras = ["a", "a+a", "a+a+a"]
for palabra in palabras:
  print("📘 Derivación descendente:")
  resultado = derivar("E", gramatica, palabra)
  for i, paso in enumerate(resultado):
      print(f"Paso {i+1}: {''.join(paso)}")
````

**Porque los agoritmos propuestos no funcionan bien con la Gramatica elegida**

**Algoritmo derivar**

Algoritmo derivar:
El algoritmo derivar no funciona correctamente debido a:

**Comprobación de longitud estricta:**

    La condición len(nueva) <= len(palabra) es muy restrictiva y no permite la
    expansion en algunos casos de palabras con simbolso terminales épsilon
    (donde un simbolo puede sustituirse por "nada", con lo que le largo final
    de la palabra "disminuye").

**Algoritmo parse_ascendente**

Algoritmo parse_ascendente_funcional:
Este algoritmo no funciona correctamente con la gramática dada debido a problemas de diseño:

**Manejo Producciones Épsilon:**

    El algoritmo no maneja una lógica para manejar la "reducción" de
    épsilon (como E' -> ε), lo cual es necesario para esta
    gramática.

**Falta Lookahead:**

    Un parser ascendente necesita "mirar" el siguiente símbolo de entrada para
    tomar decisiones correctas. El algoritmo diseñado no tiene esta
    capacidad, lo que lleva a decisiones incorrectas.

**Ausencia de Tabla de Parsing:**

    Los parsers ascendentes "avanzados" se basan en tablas de parsing que
    definen las acciones para cada estado y símbolo de entrada. La implementacion
    que venimos analizando carecede esa estructura.
    
**Orden de Reducciones:**

    La forma en que se realizan las reducciones no asegura que se encuentre
    la "reducción correcta" en el "momento adecuado", especialmente cuando
    aparecen producciones épsilon.
