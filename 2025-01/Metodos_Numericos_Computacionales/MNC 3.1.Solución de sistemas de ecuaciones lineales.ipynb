##### **Ejemplo 1: Resolución de un sistema de ecuaciones lineales**


**Descripción del problema:**

Considere el sistema de ecuaciones lineales representado por la matriz

$$A = \left[
\begin{array}{rrr|r}
4 & -1 & 2 \\
0 & -3 & 1  \\
-1 & 2 & 5  
\end{array}
\right]$$

y el vector

$$b =\left[
\begin{array}{c}
15  \\
6  \\
11  
\end{array}
\right]$$

en la ecuación $Ax = b$. Nuestro objetivo es encontrar la solución exacta del sistema utilizando Python.

**Objetivo:**
- Implementar la resolución del sistema usando la función `numpy.linalg.solve`.
- Verificar la solución obtenida.

**Salidas:**
1. El vector solución $x$.


**Ejemplo de salida esperada:**

$\text{Solución}: [ 2. -1.  3.]$

````python
import numpy as np

# Definimos la matriz de coeficientes A
A = np.array([
    [4, -1, 2],
    [0, -3, 1],
    [-1, 2, 5]
], dtype=float)

# Definimos el vector de términos independientes b
b = np.array([15, 6, 11], dtype=float)

# Resolvemos el sistema de ecuaciones Ax = b
x = np.linalg.solve(A, b)

# Mostramos la solución obtenida
print("Solución:", x)
````

##### **Ejercicio 4 (difícil): Transformación a matriz estrictamente diagonalmente dominante**

**Descripción del problema:**

Determine una nueva matriz $B$ a partir de $A$ que sea estrictamente diagonalmente dominante mediante el reordenamiento de las filas. Una matriz es estrictamente diagonalmente dominante si, en cada fila, el valor absoluto del elemento en la diagonal principal es mayor que la suma de los valores absolutos de todos los demás elementos de esa misma fila. **Tarea adicional**: piense e implemente una estrategia para realizar este ordenamiento para cualquier matriz $A$.

**Objetivo:**
- Reorganizar las filas de la matriz para hacerla estrictamente diagonalmente dominante.

**Salidas:**
1. Matriz $B$ reordenada.

````python
import numpy as np

def es_diagonalmente_dominante(matriz):
    # Extrae la diagonal principal de la matriz y aplica valor absoluto
    diagonal = np.diag(np.abs(matriz))

    # Calcula la suma de los valores absolutos de cada fila y resta el elemento diagonal
    suma_filas = np.sum(np.abs(matriz), axis=1) - diagonal

    # Verifica si TODOS los elementos diagonales son mayores que las sumas calculadas
    return np.all(diagonal > suma_filas)

def hacer_diagonalmente_dominante(A):
    n = A.shape[0]

    # Verificar si ya es diagonalmente dominante
    if es_diagonalmente_dominante(A):
        return A.copy()

    # Generar todas las permutaciones posibles de filas
    from itertools import permutations
    for perm in permutations(range(n)): # Genera todas las posibles permutaciones de los valores del
                                        # rango de A (Ej: para A 3x3 genera todas las posibles combinaciones
                                        # de los numeros (1,2,3)) e itera en esas combinaciones
        B = A[list(perm)]               # para la iteracion dada ordena las filas como indica la combinacion
        if es_diagonalmente_dominante(B):  # calcula si la cambinacion hallada es diagonalmente dominante
            return B

    print("No se encontró un reordenamiento que haga la matriz diagonalmente dominante")
    return None

# Matriz de ejemplo
A = np.array([
    [0, 2, -1],
    [1, 1, 4],
    [6, -2, 3]
], dtype=float)


print("Matriz A original:")
print(A)

# Intentamos convertirla en diagonalmente dominante
B = hacer_diagonalmente_dominante(A)

if B is not None:
    print("\nMatriz B (diagonalmente dominante):")
    print(B)
    print("\nVerificación:")
    print("¿Es B diagonalmente dominante?", es_diagonalmente_dominante(B))
````

##### **Ejercicio 5: Cálculo de una secuencia definida por una matriz**

**Descripción del problema:**

Calcule los términos de la secuencia definida por $A_n = A^n - 1/2*I$, donde $I$ es la matriz identidad, recibiendo como entrada el valor de $n$.

**Objetivo:**
- Calcular la secuencia definida a partir de la matriz $A$.

**Entradas:**
1. Número natural $n$.

**Salidas:**
1. Matriz resultante $A_n$.

````python
import numpy as np

def calcular_secuencia_A(A, n):
    error = ""
    if n < 0:
        error = "n debe ser un número natural (n ≥ 0)"

    # Verificar que A sea cuadrada e invertible
    if A.shape[0] != A.shape[1]:
        error = "La matriz A debe ser cuadrada"
    if np.linalg.det(A) == 0:
        error = "La matriz A debe ser invertible"

    I = np.eye(A.shape[0])  # Matriz identidad
    secuencia = []

    for k in range(n + 1):
        A_pow = np.linalg.matrix_power(A, k)  # Calcula A^k
        An = A_pow -((1/2)* I)              # Divide por 2*I (equivale a (1/2)*(A^k)^-1)
        secuencia.append(An)

    return secuencia,error

# Ejemplo de uso

# Definimos la matriz A
A = np.array([
    [0, 2, -1],
    [1, 1, 4],
    [6, -2, 3]
], dtype=float)

n = int(input("Ingrese el valor de n (≥ 0): "))

# Calculamos la secuencia
secuencia,error = calcular_secuencia_A(A, n)

# Mostramos resultados
if error == "":
  k=n
  print(f"\nA{n} =")
  print(np.round(secuencia[n], 4))  # Redondeamos a 4 decimales para mejor visualización
else:
  print(error)
````

#### **Métodos Iterativos: método de Jacobi**

Los "Métodos Iterativos" constituyen una clase de técnicas numéricas para resolver sistemas de ecuaciones lineales, en las cuales la solución exacta $x$ se obtiene como el límite de una secuencia de aproximaciones $x^{(0)}, x^{(1)}, \ldots$. Estos métodos son particularmente útiles cuando las matrices de los sistemas poseen características especiales, como ser grandes y dispersas (es decir, contener muchos elementos nulos).

A diferencia de los métodos directos, que buscan la solución exacta en un número fijo de pasos, los métodos iterativos construyen aproximaciones sucesivas, reduciendo el costo computacional en cada etapa.

En esta sección, estudiaremos el método de Jacobi, una técnica iterativa ampliamente utilizada.


##### **Ejemplo 1: Aplicación del método de Jacobi**



**Descripción del problema:**

Apliquemos el método de Jacobi al siguiente sistema de ecuaciones:
$$
\left\{
\begin{array}{rcc}
4x_1  -x_2  +2x_3 & = & 15\\
  -3x_2  +x_3 & = & 6\\
-x_1  +2x_2  +5x_3 & = & 11
\end{array}
\right.
$$

y calculemos las 4 primeras aproximaciones $x^{(1)}, x^{(2)}, x^{(3)}, x^{(4)}$ tomando como aproximación inicial $x^{(0)}$ el vector nulo.

**Objetivo:**
- Implementar el método de Jacobi iterativamente.
- Comparar las aproximaciones con la solución exacta.

**Salidas:**
1. Aproximaciones $x^{(1)}, x^{(2)}, x^{(3)}, x^{(4)}$.

````python
import numpy as np

A = np.array([
    [4, -1, 2],
    [0, -3, 1],
    [-1, 2, 5]
], dtype=float)

b = np.array([15, 6, 11], dtype=float)

# Aproximación inicial
x = np.zeros_like(b, dtype=float)

# Número de iteraciones
num_iteraciones = 400

# Matriz diagonal D
D = np.diag(np.diag(A))
D_inv = np.linalg.inv(D)
N = D - A

# Iteraciones de Jacobi
for k in range(num_iteraciones):
    x = np.dot(D_inv, np.dot(N, x) + b)
    print(f"Iteración {k+1}: {x}")
    resultado = np.dot(A, x)
    solucion = True
    for i in range(len(b)):
        if b[i] != resultado[i]:
            solucion = False
    if solucion :
        break
````

En la recta real, la distancia entre dos puntos se mide utilizando el valor absoluto. En espacios reales de dimensión mayor, la noción de distancia se generaliza mediante la norma euclidiana, es decir, si $x=(x_1, \ldots, x_n)$ e $y=(y_1, \ldots,y_n) \in \mathbb{R}^n $ entonces

$$||x-y|| = \left(\sum_{i=1}^n(x_i-y_i)^2\right)^{1/2}. $$

Esta medida nos permite evaluar la magnitud de un vector y su proximidad a otros puntos en el espacio. A continuación, mostramos cómo calcular esta norma usando Python.

##### **Ejercicio 1: Implementación del método de Jacobi con criterio de convergencia**

**Descripción del problema:**

Implemente una función en Python que aplique el método de Jacobi para resolver el sistema de ecuaciones
$$
\left\{
\begin{array}{rcc}
5x_1  -2x_2  +3x_3 & = & 8\\
3x_1   +8x_2  -4x_3 & = & 10\\
x_1  -5x_2  +7x_3 & = & 6
\end{array}
\right.
$$
hasta alcanzar una tolerancia de $10^{-4}$.

**Objetivo:**
- Implementar iteraciones hasta la convergencia utilizando la norma euclidiana.

**Salidas:**
1. Vector solución aproximada.



````python
import numpy as np

def metodo_jacobi(A, b, tol=1e-4, max_iter=1000):
    # Resuelve el sistema Ax = b usando el método de Jacobi.
    # Args:
    #    A (np.ndarray): Matriz de coeficientes (n x n)
    #    b (np.ndarray): Vector de términos independientes (n,)
    #    tol (float): Tolerancia para el criterio de parada
    #    max_iter (int): Número máximo de iteraciones permitidas
    #Retorna:
    #    np.ndarray: Vector solución x
    #    int: Número de iteraciones realizadas

    n = len(b)
    x = np.zeros(n)  # Solución inicial (vector cero)
    x_nuevo = np.zeros(n)
    iteraciones = 0
    D = np.diag(np.diag(A))
    D_inv = np.linalg.inv(D)
    N = D - A


    while np.linalg.norm(b-np.dot(A,x)) >= tol:
        x_nuevo = np.dot(D_inv, np.dot(N, x) + b)
        # print(f"Iteración {iteraciones+1}: {x}")
        if iteraciones >= max_iter:
            break


        # Criterio de convergencia (norma euclidiana del cambio relativo)
        error = np.linalg.norm(x_nuevo - x) / np.linalg.norm(x_nuevo)
        if error < tol:
            break

        x = x_nuevo.copy()
        iteraciones += 1

    return x_nuevo, iteraciones

# Sistema de ecuaciones dado
A = np.array([
    [5, -2, 3],
    [3, 8, -4],
    [1, -5, 7]
], dtype=float)

b = np.array([8, 10, 6], dtype=float)

# Resolver usando Jacobi
error=1e-4
solucion, iteraciones = metodo_jacobi(A, b,error,1000)

print(f"Solución aproximada: {solucion}")
print(f"Número de iteraciones: {iteraciones}")
print(f"Verificación (A . solucion): {np.dot(A, solucion)}")
````

#### **Métodos Iterativos: método de Gauss-Seidel**

A diferencia del método de Jacobi, el método de Gauss-Seidel utiliza los valores ya calculados de las componentes de la solución en la misma iteración para acelerar la convergencia.

##### **Ejemplo: Aplicación del método de Gauss-Seidel**

**Descripción del problema:**

Resolver el siguiente sistema de ecuaciones:
$$
\left\{
\begin{array}{rcc}
4x_1  -x_2  +2x_3 & = & 15\\
  -3x_2  +x_3 & = & 6\\
-x_1  +2x_2  +5x_3 & = & 11
\end{array}
\right.
$$

**Objetivo:**
- Implementar el método de Gauss-Seidel.
- Obtener aproximaciones sucesivas de la solución.

**Salidas:**
1. Aproximaciones de la solución en cada iteración.

````python
import numpy as np

def gauss_seidel(A, b, x0, max_iter=10, tol=1e-4):
    """
    Implementa el método de Gauss-Seidel para resolver sistemas lineales Ax = b.

    Parámetros:
    A (ndarray): Matriz de coeficientes (n x n).
    b (ndarray): Vector de términos independientes (n).
    x0 (ndarray): Vector inicial para las iteraciones (n).
    max_iter (int): Número máximo de iteraciones.
    tol (float): Tolerancia para el criterio de convergencia.

    Retorna:
    ndarray: Vector solución aproximada.
    int: Número de iteraciones realizadas.
    """
    n = len(b)  # Número de ecuaciones
    x = x0.copy()  # Inicializamos el vector solución con la aproximación inicial

    for k in range(max_iter):
        x_new = np.copy(x)  # Copia del vector solución para actualizar en la iteración

        for i in range(n):
            # Suma de los términos antes de la diagonal principal
            sum1 = np.dot(A[i, :i], x_new[:i])

            # Suma de los términos después de la diagonal principal
            sum2 = np.dot(A[i, i+1:], x[i+1:])

            # Actualizamos el valor de x[i] usando la fórmula del método de Gauss-Seidel
            x_new[i] = (b[i] - sum1 - sum2) / A[i, i]

        # Verificamos la convergencia usando la norma infinito
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k + 1

        x = x_new  # Actualizamos el vector solución para la siguiente iteración
        print(f"Solucion encontrada: {x}, Iteracion : {k}")

    return x, max_iter  # Retornamos la última solución y el número máximo de iteraciones

# Definimos la matriz de coeficientes A y el vector de términos independientes b
A = np.array([[4, -1, 2], [0, -3, 1], [-1, 2, 5]], dtype=float)
b = np.array([15, 6, 11], dtype=float)

# Vector inicial (x0) con todos los valores en cero
x0 = np.zeros_like(b)

# Llamamos a la función para resolver el sistema
sol, iteraciones = gauss_seidel(A, b, x0)

# Mostramos la solución aproximada y el número de iteraciones realizadas
print(f"Solución aproximada: {sol}")
print(f"Iteraciones: {iteraciones}")
````
