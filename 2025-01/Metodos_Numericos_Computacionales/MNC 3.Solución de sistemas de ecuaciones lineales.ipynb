##### **Ejemplo 1: Resolución de un sistema de ecuaciones lineales**


**Descripción del problema:**

Considere el sistema de ecuaciones lineales representado por la matriz

$$A = \left[
\begin{array}{rrr|r}
4 & -1 & 2 \\
0 & -3 & 1  \\
-1 & 2 & 5  
\end{array}
\right]$$

y el vector

$$b =\left[
\begin{array}{c}
15  \\
6  \\
11  
\end{array}
\right]$$

en la ecuación $Ax = b$. Nuestro objetivo es encontrar la solución exacta del sistema utilizando Python.

**Objetivo:**
- Implementar la resolución del sistema usando la función `numpy.linalg.solve`.
- Verificar la solución obtenida.

**Salidas:**
1. El vector solución $x$.


**Ejemplo de salida esperada:**

$\text{Solución}: [ 2. -1.  3.]$

````python
import numpy as np

# Definimos la matriz de coeficientes A
A = np.array([
    [4, -1, 2],
    [0, -3, 1],
    [-1, 2, 5]
], dtype=float)

# Definimos el vector de términos independientes b
b = np.array([15, 6, 11], dtype=float)

# Resolvemos el sistema de ecuaciones Ax = b
x = np.linalg.solve(A, b)

# Mostramos la solución obtenida
print("Solución:", x)
````

##### **Ejercicio 4 (difícil): Transformación a matriz estrictamente diagonalmente dominante**

**Descripción del problema:**

Determine una nueva matriz $B$ a partir de $A$ que sea estrictamente diagonalmente dominante mediante el reordenamiento de las filas. Saiba que una matriz es estrictamente diagonalmente dominante si el módulo de cada elemento de la diagonal principal es mayor que la suma de los módulos de los elementos de su fila. **Tarea adicional**: piense e implemente una estrategia para realizar este ordenamiento para cualquier matriz $A$.

**Objetivo:**
- Reorganizar las filas de la matriz para hacerla estrictamente diagonalmente dominante.

**Salidas:**
1. Matriz $B$ reordenada.

````python
import numpy as np

# Función para verificar si la matriz es estrictamente diagonalmente dominante
def es_diagonalmente_dominante(A):
    filas, columnas = A.shape
    suma_diagonal = 0
    for i in range(filas):
        suma_no_diagonal =0
        for j in range(filas):
          if i != j:
            suma_no_diagonal = suma_no_diagonal + abs(A[i,j])
        suma_diagonal = abs(A[i,i])
        if suma_diagonal <= suma_no_diagonal:
            return False
    return True

# Función para reordenar las filas de la matriz para que sea estrictamente diagonalmente dominante
def reordenar_matriz(A):
    filas, columnas = A.shape
    B = A.copy()  # Hacemos una copia de A para no modificarla directamente

    # Intentamos reordenar las filas para hacer la matriz diagonalmente dominante
    for i in range(filas):
        # Seleccionamos la fila con el valor absoluto más grande en la columna i (sin considerar la diagonal)
        max_fila_idx = i
        max_valor = np.abs(B[i, i])

        for j in range(i + 1, filas):
            if np.abs(B[j, i]) > max_valor:
                max_fila_idx = j
                max_valor = np.abs(B[j, i])

        # Si encontramos una fila con un mayor valor absoluto, intercambiamos las filas
        if max_fila_idx != i:
            B[[i, max_fila_idx], :] = B[[max_fila_idx, i], :]

    # Si la matriz reordenada es diagonalmente dominante, devolvemos la matriz resultante
    if es_diagonalmente_dominante(B):
        return B
    else:
        raise ValueError("No es posible hacer la matriz estrictamente diagonalmente dominante con el reordenamiento de filas.")

# Ejemplo de uso
A = np.array([
    [0, 2, -1],
    [1, 1, 4],
    [6, -2, 3]
], dtype=float)

print("Matriz original A:")
print(A)

# Reordenamos la matriz para hacerla estrictamente diagonalmente dominante
try:
    B = reordenar_matriz(A)
    print("\nMatriz B (reordenada y estrictamente diagonalmente dominante):")
    print(B)
except ValueError as e:
    print("\nError:", e)
````

#### **Métodos Iterativos: método de Jacobi**

Los "Métodos Iterativos" constituyen una clase de técnicas numéricas para resolver sistemas de ecuaciones lineales, en las cuales la solución exacta $x$ se obtiene como el límite de una secuencia de aproximaciones $x^{(0)}, x^{(1)}, \ldots$. Estos métodos son particularmente útiles cuando las matrices de los sistemas poseen características especiales, como ser grandes y dispersas (es decir, contener muchos elementos nulos).

A diferencia de los métodos directos, que buscan la solución exacta en un número fijo de pasos, los métodos iterativos construyen aproximaciones sucesivas, reduciendo el costo computacional en cada etapa.

En esta sección, estudiaremos el método de Jacobi, una técnica iterativa ampliamente utilizada.


##### **Ejemplo 1: Aplicación del método de Jacobi**



**Descripción del problema:**

Apliquemos el método de Jacobi al siguiente sistema de ecuaciones:
$$
\left\{
\begin{array}{rcc}
4x_1  -x_2  +2x_3 & = & 15\\
  -3x_2  +x_3 & = & 6\\
-x_1  +2x_2  +5x_3 & = & 11
\end{array}
\right.
$$

y calculemos las 4 primeras aproximaciones $x^{(1)}, x^{(2)}, x^{(3)}, x^{(4)}$ tomando como aproximación inicial $x^{(0)}$ el vector nulo.

**Objetivo:**
- Implementar el método de Jacobi iterativamente.
- Comparar las aproximaciones con la solución exacta.

**Salidas:**
1. Aproximaciones $x^{(1)}, x^{(2)}, x^{(3)}, x^{(4)}$.

````python
import numpy as np

A = np.array([
    [4, -1, 2],
    [0, -3, 1],
    [-1, 2, 5]
], dtype=float)

b = np.array([15, 6, 11], dtype=float)

# Aproximación inicial
x = np.zeros_like(b, dtype=float)

# Número de iteraciones
num_iteraciones = 4

# Matriz diagonal D
D = np.diag(np.diag(A))
D_inv = np.linalg.inv(D)
N = D - A

# Iteraciones de Jacobi
for k in range(num_iteraciones):
    x = np.dot(D_inv, np.dot(N, x) + b)
    print(f"Iteración {k+1}: {x}")
````

En la recta real, la distancia entre dos puntos se mide utilizando el valor absoluto. En espacios reales de dimensión mayor, la noción de distancia se generaliza mediante la norma euclidiana, es decir, si $x=(x_1, \ldots, x_n)$ e $y=(y_1, \ldots,y_n) \in \mathbb{R}^n $ entonces

$$||x-y|| = \left(\sum_{i=1}^n(x_i-y_i)^2\right)^{1/2}. $$

Esta medida nos permite evaluar la magnitud de un vector y su proximidad a otros puntos en el espacio. A continuación, mostramos cómo calcular esta norma usando Python.

#### **Métodos Iterativos: método de Gauss-Seidel**

A diferencia del método de Jacobi, el método de Gauss-Seidel utiliza los valores ya calculados de las componentes de la solución en la misma iteración para acelerar la convergencia.

##### **Ejemplo: Aplicación del método de Gauss-Seidel**

**Descripción del problema:**

Resolver el siguiente sistema de ecuaciones:
$$
\left\{
\begin{array}{rcc}
4x_1  -x_2  +2x_3 & = & 15\\
  -3x_2  +x_3 & = & 6\\
-x_1  +2x_2  +5x_3 & = & 11
\end{array}
\right.
$$

**Objetivo:**
- Implementar el método de Gauss-Seidel.
- Obtener aproximaciones sucesivas de la solución.

**Salidas:**
1. Aproximaciones de la solución en cada iteración.

````python
import numpy as np

def gauss_seidel(A, b, x0, max_iter=10, tol=1e-4):
    """
    Implementa el método de Gauss-Seidel para resolver sistemas lineales Ax = b.

    Parámetros:
    A (ndarray): Matriz de coeficientes (n x n).
    b (ndarray): Vector de términos independientes (n).
    x0 (ndarray): Vector inicial para las iteraciones (n).
    max_iter (int): Número máximo de iteraciones.
    tol (float): Tolerancia para el criterio de convergencia.

    Retorna:
    ndarray: Vector solución aproximada.
    int: Número de iteraciones realizadas.
    """
    n = len(b)  # Número de ecuaciones
    x = x0.copy()  # Inicializamos el vector solución con la aproximación inicial

    for k in range(max_iter):
        x_new = np.copy(x)  # Copia del vector solución para actualizar en la iteración

        for i in range(n):
            # Suma de los términos antes de la diagonal principal
            sum1 = np.dot(A[i, :i], x_new[:i])

            # Suma de los términos después de la diagonal principal
            sum2 = np.dot(A[i, i+1:], x[i+1:])

            # Actualizamos el valor de x[i] usando la fórmula del método de Gauss-Seidel
            x_new[i] = (b[i] - sum1 - sum2) / A[i, i]

        # Verificamos la convergencia usando la norma infinito
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k + 1

        x = x_new  # Actualizamos el vector solución para la siguiente iteración

    return x, max_iter  # Retornamos la última solución y el número máximo de iteraciones

# Definimos la matriz de coeficientes A y el vector de términos independientes b
A = np.array([[4, -1, 2], [0, -3, 1], [-1, 2, 5]], dtype=float)
b = np.array([15, 6, 11], dtype=float)

# Vector inicial (x0) con todos los valores en cero
x0 = np.zeros_like(b)

# Llamamos a la función para resolver el sistema
sol, iteraciones = gauss_seidel(A, b, x0)

# Mostramos la solución aproximada y el número de iteraciones realizadas
print(f"Solución aproximada: {sol}")
print(f"Iteraciones: {iteraciones}")
````
