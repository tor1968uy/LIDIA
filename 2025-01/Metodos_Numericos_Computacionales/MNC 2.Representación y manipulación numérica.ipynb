#### **Convergencia Numérica**

En la práctica, muchas veces es difícil determinar exactamente el límite de una secuencia numérica. Intuitivamente, podemos pensar que es razonable aproximar ese límite mediante un elemento de la secuencia, siempre que los elementos siguientes no presenten variaciones significativas.

¡La buena noticia es que esta intuición es correcta!

Matemáticamente, sabemos que, si $(x_n)$ es una secuencia de números reales convergente, es decir, si $\exists L \in \mathbb{R}$ tal que $x_n \to L$ cuando $n \to \infty$, entonces, para cualquier $\varepsilon > 0$, existe un índice $n_0$ tal que $|x_n - L| < \varepsilon$ para todo $n > n_0$. Además, el error absoluto al aproximar $L$ mediante $x_{n_0}$ será menor que $\varepsilon$, lo que garantiza la precisión deseada.

Por lo tanto, el control del error es un paso crucial en la práctica numérica, ya que nos permite determinar cuándo una aproximación es suficientemente cercana al valor verdadero.

##### **Ejemplo: Control del error en una secuencia convergente**



**Descripción del problema:**

Considere la secuencia definida por $x_n = \frac{1}{n}$. El límite de esta secuencia es $L = 0$, ya que $\frac{1}{n} \to 0$ cuando $n \to \infty$.

Escriba un programa en Python que permita calcular el menor índice $n_0$ tal que el error absoluto $|x_n - L|$ sea menor que un valor de tolerancia $\varepsilon$ proporcionado por el usuario.

**Objetivo:**
- Implementar la fórmula cerrada de la secuencia $x_n = \frac{1}{n}$.
- Determinar el menor $n_0$ para un dado $\varepsilon > 0$.
- Confirmar que $|x_{n_0} - L| < \varepsilon$.

**Entradas:**
1. Valor de $\varepsilon$ (tolerancia del error absoluto).

**Salidas:**
1. El índice $n_0$ encontrado.
2. El valor de $x_{n_0}$.
3. El error absoluto $|x_{n_0} - L|$.


**Ejemplo de entrada y salida esperada:**

Entrada:
$$ \varepsilon = 0.01 $$

Salida esperada:
\begin{aligned}
    n_0 &= 100 \\
    x_{n_0} &= 0.01 \\
    |x_{n_0} - L| &= 0.01
\end{aligned}

**Tarea adicional:**

Experimente con diferentes valores de $\varepsilon$ y observe cómo cambia el índice $n_0$.

```python
# Función para encontrar el menor n_0 tal que |x_n - L| < ε
def encontrar_n0(epsilon):
    # Fórmula cerrada de la secuencia
    limite = 0  # Límite conocido
    n = 1  # Inicializamos n desde 1

    while True:
        x_n = 1 / n  # Calculamos el término de la secuencia
        error = abs(x_n - limite)  # Calculamos el error absoluto

        if error < epsilon:  # Comprobamos si cumple con la tolerancia
            return n, x_n, error

        n += 1  # Incrementamos n si no cumple la condición

# Entrada del usuario
epsilon = float(input("Ingrese el valor de ε (tolerancia): "))

# Llamamos a la función
n0, x_n0, error = encontrar_n0(epsilon)

# Mostramos los resultados
print(f"El menor n_0 encontrado es: {n0}")
print(f"El valor de x_n0 es: {x_n0:.5f}")
print(f"El error absoluto es: {error:.5f}")
```

##### **Ejercicio 1: Control de errores sin conocer el límite**

**Descripción del problema:**

Considere la secuencia definida por $x_n = \left(1+\frac{1}{n}\right)^n$. Se sabe que se trata de una secuencia convergente, pero esta vez el límite no se dará explícitamente.

Escriba un programa en Python que calcule el menor índice $n_0$ tal que la diferencia absoluta entre dos términos consecutivos de la secuencia, $ ∣x_{n+1}−x_n∣ $, sea menor que un valor de tolerancia $\varepsilon$ proporcionado por el usuario.

**Objetivo:**
- Implementar la fórmula cerrada de la secuencia $x_n = \left(1+\frac{1}{n}\right)^n$.
- Determinar $n_0$ para un dado $\varepsilon >0$ sin conocer explícitamente el límite $L$.
- Basar la decisión de convergencia únicamente en la diferencia entre términos consecutivos en la secuencia.

**Entradas:**
1. Valor de $\varepsilon$ (tolerancia del error entre términos consecutivos).

**Salidas:**
1. El índice $n_0$ encontrado.
2. El valor de $x_{n_0}$.
3. El error absoluto $|x_{n_0+1} - x_{n_0}|$.


**Ejemplo de entrada y salida esperada:**

Entrada:
$$ \varepsilon = 0.001 $$

Salida esperada (aproximada):
\begin{aligned}
    n_0 &= 1000 \\
    x_{n_0} &= 2.71692 \\
    |x_{n_0+1} - x_{n_0}| &= 0.00099.
\end{aligned}

```python

## Ejercicio 1
# ##########################################################################################################


# Función para encontrar el menor n_0 tal que |x_{n+1} - x_n| < epsilon
def encontrar_n0(epsilon):
    # Inicializamos n y calculamos el primer término de la secuencia
    n = 1
    x_n = (1 + 1 / n) ** n  # Calculamos x_n

    # Bucle infinito hasta que se cumpla la condición
    while True:
        n += 1  # Incrementamos n
        x_n_mas_1 = (1 + 1 / n) ** n  # Calculamos x_{n+1}
        error = abs(x_n_mas_1 - x_n)  # Calculamos el error absoluto

        # Si el error es menor que epsilon, devolvemos los resultados
        if error < epsilon:
            return n, x_n, error

        # Actualizamos x_n para la siguiente iteración
        x_n = x_n_mas_1


# Entrada del usuario para la tolerancia epsilon
epsilon = float(input("Ingrese el valor de ε (tolerancia): "))

# Llamamos a la función para encontrar n_0
n0, x_n0, error = encontrar_n0(epsilon)

# Mostramos los resultados
print(f"El menor n_0 encontrado es: {n0}")
print(f"El valor de x_n0 es: {x_n0:.5f}")
print(f"El error absoluto es: {error:.5f}")
```

Antes de avanzar con otros ejercicios, es importante mencionar que estamos utilizando dos hipótesis fundamentales:

1. Estas secuencias son convergentes, es decir, existe un valor real al que tienden cuando $n\to \infty$.
2. Estas secuencias son uniformemente convergentes, es decir, los términos de la secuencia se acercan uniformemente al límite sin que ocurra la mala suerte de que los términos vuelvan a alejarse entre sí.

Podríamos crear ejemplos de secuencias con mal comportamiento que no cumplan estas características, pero eso queda para otra ocasión.

##### **Ejercicio 2: Secuencias recurrentes y control de error absoluto**

**Descripción del problema:**

Considere la secuencia definida por la relación recurrente:
$$ x_{n+1} = 1 + \frac{1}{x_n}, ~~~x_1 >0. $$


Se sabe que se trata de una secuencia convergente.

Escriba un programa en Python que calcule el menor índice $n_0$​ tal que la diferencia absoluta entre dos términos consecutivos de la secuencia,$ ∣x_{n+1}−x_n∣ $, sea menor que un valor de tolerancia $\varepsilon$ proporcionado por el usuario.

**Objetivo:**
- Implementar la fórmula de recurrencia $x_{n+1} = 1 + \frac{1}{x_n}$.
- Determinar $n_0$ para un dado $\varepsilon >0$ sin conocer explícitamente el límite.
- Basar la decisión de convergencia únicamente en la diferencia entre términos consecutivos en la secuencia.

**Entradas:**
1. Valor de $\varepsilon$ (tolerancia del error entre términos consecutivos).
2. Valor inicial $x_1>0​$

**Salidas:**
1. El índice $n_0$ encontrado.
2. El valor de $x_{n_0}$.
3. El error absoluto $|x_{n_0+1} - x_{n_0}|$.


**Ejemplo de entrada y salida esperada:**

Entrada:
$$ \varepsilon = 0.001, ~~~ x_1=1.5 $$

Salida esperada (aproximada):
\begin{aligned}
    n_0 &= 10 \\
    x_{n_0} &= 1.61798 \\
    |x_{n_0+1} - x_{n_0}| &= 0.00099.
\end{aligned}

```python
## Ejercicio 2
# ##########################################################################################################

# Función para encontrar el menor n_0 tal que |x_{n+1} - x_n| < epsilon
def encontrar_n0_recurrente(epsilon, x1):
    # Inicializamos n y el primer término de la secuencia
    n = 1
    x_n = x1  # x1 es el valor inicial proporcionado por el usuario

    # Bucle infinito hasta que se cumpla la condición
    while True:
        n += 1  # Incrementamos n
        x_n_mas_1 = 1 + 1 / x_n  # Calculamos x_{n+1} usando la recurrencia
        error = abs(x_n_mas_1 - x_n)  # Calculamos el error absoluto

        # Si el error es menor que epsilon, devolvemos los resultados
        if error < epsilon:
            return n, x_n, error

        # Actualizamos x_n para la siguiente iteración
        x_n = x_n_mas_1


# Entrada del usuario para la tolerancia epsilon y el valor inicial x1
epsilon = float(input("Ingrese el valor de ε (tolerancia): "))
x1 = float(input("Ingrese el valor inicial x1 (> 0): "))

# Llamamos a la función para encontrar n_0
n0, x_n0, error = encontrar_n0_recurrente(epsilon, x1)

# Mostramos los resultados
print(f"El menor n_0 encontrado es: {n0}")
print(f"El valor de x_n0 es: {x_n0:.5f}")
print(f"El error absoluto es: {error:.5f}")
```

##### **Ejercicio 3: Secuencia híbrida y control de error relativo**

**Descripción del problema:**

Considere la secuencia definida por:
$$ x_1 =2, ~~~~ x_{n+1} = x_n + \frac{(-1)^n}{n^2}, ~n\geqslant 1. $$


Se sabe que esta secuencia converge a un valor finito.

Escriba un programa en Python que calcule el menor índice $n_0$​ tal que el error relativo entre dos términos consecutivos de la secuencia,$ \frac{∣x_{n+1}−x_n∣}{|x_n|} $, sea menor que un valor de tolerancia $\varepsilon$ proporcionado por el usuario.

**Objetivo:**
- Implementar una fórmula híbrida de secuencia.
- Determinar $n_0$ para un dado $\varepsilon >0$.
- Introducir el concepto de oscilaciones en la convergencia debido a los términos alternantes.

**Entradas:**
1. Valor de $\varepsilon$ (tolerancia del error relativo).

**Salidas:**
1. El índice $n_0$ encontrado.
2. El valor de $x_{n_0}$.
3. El error relativo  $\frac{|x_{n_0+1} - x_{n_0}|}{|x_{n_0}|}$.


**Ejemplo de entrada y salida esperada:**

Entrada:
$$ \varepsilon = 0.001 $$

Salida esperada (aproximada):
\begin{aligned}
    n_0 &= 93 \\
    x_{n_0} &= 1.17748 \\
    \frac{|x_{n_0+1} - x_{n_0}|}{|x_{n_0}|} &= 0.00010.
\end{aligned}

**Tarea adicional:**

Experimente utilizar $x_1 =1$. ¿Qué sucede? Explique una solución.

```python
## Ejercicio 3
# ##########################################################################################################

# Función para encontrar el menor n_0 tal que |(x_{n+1} - x_n) / x_n| < epsilon
def encontrar_n0_hibrida(epsilon):
    # Inicializamos n y el primer término de la secuencia
    n = 1
    x_n = 1  # x1 es 2 según la descripción del problema

    # Bucle infinito hasta que se cumpla la condición
    while True:
        n += 1  # Incrementamos n
        x_n_mas_1 = x_n + ((-1) ** n) / (n ** 2)  # Calculamos x_{n+1}
        error_relativo = abs((x_n_mas_1 - x_n) / x_n)  # Calculamos el error relativo

        # Si el error relativo es menor que epsilon, devolvemos los resultados
        if error_relativo < epsilon:
            return n, x_n, error_relativo

        # Actualizamos x_n para la siguiente iteración
        x_n = x_n_mas_1


# Entrada del usuario para la tolerancia epsilon
epsilon = float(input("Ingrese el valor de ε (tolerancia): "))

# Llamamos a la función para encontrar n_0
n0, x_n0, error_relativo = encontrar_n0_hibrida(epsilon)

# Mostramos los resultados
print(f"El menor n_0 encontrado es: {n0}")
print(f"El valor de x_n0 es: {x_n0:.5f}")
print(f"El error relativo es: {error_relativo:.5f}")
```

##### **Ejercicio 4: Aproximación de $\pi$ mediante una secuencia convergente**

**Descripción del problema:**

Considere la secuencia dada por:
$$ x_{n} = 4\cdot \sum_{k=0}^{n}\frac{(-1)^k}{2k+1}, ~n\geqslant 0. $$


Esta secuencia está basada en la serie de Leibniz para la aproximación del número $\pi$. Sabemos que $x_n\to\pi$ cuando $n\to\infty$.

Escriba un programa en Python que calcule el menor índice $n_0$​ tal que el error absoluto entre dos términos consecutivos de la secuencia, $∣x_{n+1}−x_n∣$, sea menor que un valor de tolerancia $\varepsilon$ proporcionado por el usuario.

**Objetivo:**
- Implementar la fórmula cerrada de una secuencia convergente relacionada con $\pi$.
- Determinar $n_0$ en función de un error absoluto especificado.
- Familiarizarse con sumas parciales y su relación con el límite.

**Entradas:**
1. Valor de $\varepsilon$ (tolerancia del error absoluto entre términos consecutivos).

**Salidas:**
1. El índice $n_0$ encontrado.
2. El valor de $x_{n_0}$.
3. El error relativo  $|x_{n_0+1} - x_{n_0}|$.


**Ejemplo de entrada y salida esperada:**

Entrada:
$$ \varepsilon = 0.001 $$

Salida esperada (aproximada):
\begin{aligned}
    n_0 &= 2000 \\
    x_{n_0} &= 3.14209 \\
    |x_{n_0+1} - x_{n_0}| &= 0.001.
\end{aligned}

```python
## Ejercicio 4
# ##########################################################################################################

# Función para encontrar el menor n_0 tal que |x_{n+1} - x_n| < epsilon
def encontrar_n0_pi(epsilon):
    # Inicializamos n y la suma parcial de la serie
    n = 0
    suma_parcial = 0  # Inicializamos la suma parcial en 0

    # Bucle infinito hasta que se cumpla la condición
    while True:
        # Calculamos el término actual de la serie
        termino = ((-1) ** n) / (2 * n + 1)
        suma_parcial += termino  # Actualizamos la suma parcial

        # Calculamos x_n y x_{n+1}
        x_n = 4 * suma_parcial
        n += 1  # Incrementamos n
        termino_siguiente = ((-1) ** n) / (2 * n + 1)
        x_n_mas_1 = 4 * (suma_parcial + termino_siguiente)

        # Calculamos el error absoluto
        error = abs(x_n_mas_1 - x_n)

        # Si el error es menor que epsilon, devolvemos los resultados
        if error < epsilon:
            return n, x_n, error


# Entrada del usuario para la tolerancia epsilon
epsilon = float(input("Ingrese el valor de ε (tolerancia): "))

# Llamamos a la función para encontrar n_0
n0, x_n0, error = encontrar_n0_pi(epsilon)

# Mostramos los resultados
print(f"El menor n_0 encontrado es: {n0}")
print(f"El valor de x_n0 es: {x_n0:.5f}")
print(f"El error absoluto es: {error:.5f}")
```
