#### **Interpolación polinómica**

La interpolación polinómica es una técnica que nos permite construir un polinomio que pasa exactamente por un conjunto de puntos dados. Este polinomio puede ser usado para estimar valores intermedios entre los puntos o extrapolar fuera de ellos, aunque esto último puede tener menor precisión.

##### **Ejemplo: Interpolación lineal**

**Descripción del problema:**

Considere los puntos $A = {(1, 2), (3, 4)}$. Encuentre el polinomio de primer grado $P(x)$ que interpola estos puntos. Además, grafique el polinomio y los puntos dados.

**Objetivo:**
- Calcular el polinomio de primer grado.
- Visualizar el polinomio y los puntos en un gráfico.

**Salidas:**
1. Expresión del polinomio interpolador.
2. Gráfico del polinomio y los puntos.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir los puntos
x = np.array([1, 3])  # Coordenadas x
y = np.array([2, 4])  # Coordenadas y

# Calcular los coeficientes del polinomio interpolador
coeficientes = np.polyfit(x, y, deg=1)  # Grado 1 para interpolación lineal

# Crear una función polinómica a partir de los coeficientes
polinomio = np.poly1d(coeficientes)

# Puntos para graficar el polinomio
x_vals = np.linspace(0, 4, 100)

y_vals = polinomio(x_vals)

# Graficar los puntos y el polinomio
plt.scatter(x, y, color='red', label='Puntos dados')
plt.plot(x_vals, y_vals, label=f'P(x) = {coeficientes[0]:.2f}x + {coeficientes[1]:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolación lineal')
plt.legend()
plt.grid()
plt.show()
````

#### **Splines cúbicos**

En esta sección, exploraremos cómo utilizar los splines cúbicos para realizar interpolaciones más suaves y precisas. Los splines cúbicos dividen el intervalo en secciones más pequeñas y ajustan polinomios de tercer grado en cada segmento, asegurando continuidad en la función y en sus derivadas hasta el segundo orden. Este método evita las oscilaciones problemáticas de los polinomios de alto grado.



##### **Ejemplo 1: Interpolación con Splines Cúbicos Naturales**

**Descripción del problema:**

Considere los puntos $A = {(0, 0), (1, 2), (2, 0)}$. Encuentre el spline cúbico natural que interpola estos puntos. Además, grafique el spline y los puntos dados.

**Objetivo:**
- Construir el spline cúbico natural que interpola los puntos.
- Visualizar el spline y los puntos en un gráfico.

**Salidas:**
1. Expresión del spline cúbico natural.
2. Gráfico del spline y los puntos.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Definir los puntos
x = np.array([0, 1, 2])
y = np.array([0, 2, 0])

# Construir el spline cúbico natural
spline = CubicSpline(x, y, bc_type='natural')

# Mostrar la expresión analítica del spline en cada intervalo
for i in range(len(x) - 1):
    coef = spline.c[:, i]
    print(f"S_{i}(x) = {coef[0]:.3f}(x - {x[i]})^3 + {coef[1]:.3f}(x - {x[i]})^2 + "
          f"{coef[2]:.3f}(x - {x[i]}) + {coef[3]:.3f}")

# Puntos para graficar el spline
x_vals = np.linspace(0, 2, 100)
y_vals = spline(x_vals)

# Graficar los puntos y el spline
plt.scatter(x, y, color='red', label='Puntos dados')
plt.plot(x_vals, y_vals, label='Spline cúbico natural')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolación con Splines Cúbicos Naturales')
plt.legend()
plt.grid()
plt.show()
````

#### **Aproximación por mínimos cuadrados**

La aproximación por mínimos cuadrados es una técnica utilizada para encontrar una combinación lineal de funciones que mejor se ajusta a un conjunto de puntos observados. Este método es particularmente útil cuando los datos contienen ruido, inconsistencias o no siguen exactamente un patrón específico.

El objetivo es minimizar la suma de los cuadrados de las diferencias entre los valores observados y los valores ajustados por la función objetivo. Este enfoque permite construir modelos que representan de manera eficiente la tendencia general de los datos.

##### **Ejemplo: Ajuste lineal con mínimos cuadrados**

**Descripción del problema:**

Dado un conjunto de puntos $A=\{(-1,0), (0,-1), (1,0), (2,7)\}$ en un plano cartesiano, queremos encontrar la recta de la forma:

$$ F(x) = a_1 + a_2x$$

que mejor se ajuste a los datos utilizando el método de mínimos cuadrados. Luego, representaremos gráficamente los puntos junto con la recta ajustada.

**Objetivo:**
- Calcular la recta $F(x)$ que minimiza el error cuadrático con respecto a los datos dados.
- Representar los puntos y la recta ajustada en un gráfico.

**Salidas:**
1. Expresión de la recta ajustada en la forma $F(x) = a_1 + a_2x$.
2. Gráfico con los puntos y la recta ajustada.

````python
from numpy.polynomial import Polynomial
import matplotlib.pyplot as plt

# Definir los puntos
x = np.array([-1, 0, 1, 2])  # Coordenadas x
y = np.array([0, -1, 0, 7])  # Coordenadas y

# Calcular los coeficientes del ajuste lineal
coeficientes = np.polyfit(x, y, deg=1)  # Grado 1 para ajuste lineal

# Crear la función de la recta ajustada
recta_ajustada = np.poly1d(coeficientes)

# Generar valores de x para graficar la recta
x_vals = np.linspace(-2, 3, 100)
y_vals = recta_ajustada(x_vals)

# Imprimir la ecuación de la recta ajustada
print(f"La recta ajustada es: F(x) = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}")

# Graficar los puntos y la recta ajustada
plt.scatter(x, y, color='red', label='Puntos dados')
plt.plot(x_vals, y_vals, label=f'F(x) = {coeficientes[0]:.2f}x + {coeficientes[1]:.2f}', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste lineal por mínimos cuadrados')
plt.legend()
plt.grid()
plt.show()
````

##### **Ejemplo 2: Aproximación de una Función Mixta**

**Problema:**

Dado el conjunto de puntos observados:
$$ \{(1, 2.2),(2, 2.8),(3, 3.6),(4, 5.1),(5, 7.3)\}$$
ajustar una función de la forma:
$$ F(x) = a_1 + a_2x+ a_3e^x$$
usando el método de mínimos cuadrados. Luego, calcular el coeficiente de determinación $R^2$ para evaluar la calidad del ajuste.

**Objetivos:**

1. Encontrar los coeficientes $a_1$, $a_2$, $a_3$.
2. Visualizar el ajuste de la función $F(x)$ a los puntos observados.
3. Calcular el error $R^2$ para medir la calidad del ajuste.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definir los puntos observados
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.2, 2.8, 3.6, 5.1, 7.3])

# Definir la función modelo F(x) = a1 + a2*x + a3*exp(x)
def modelo(x, a1, a2, a3):
    return a1 + a2*x + a3*np.exp(x)

# Ajuste de parámetros usando mínimos cuadrados no lineales
parametros, _ = curve_fit(modelo, x_data, y_data)

# Extracción de los coeficientes ajustados
a1, a2, a3 = parametros
print(f"Coeficientes ajustados: a1 ≈ {a1:.4f}, a2 ≈ {a2:.4f}, a3 ≈ {a3:.4f}")

# Calcular los valores ajustados
y_ajustado = modelo(x_data, a1, a2, a3)

# Calcular el coeficiente de determinación R^2
y_media = np.mean(y_data)
SST = np.sum((y_data - y_media) ** 2)  # Suma total de los cuadrados
SSR = np.sum((y_data - y_ajustado) ** 2)  # Suma de los cuadrados de los residuos
R2 = 1 - (SSR / SST)

print(f"Coeficiente de determinación R² ≈ {R2:.4f}")

# Crear valores continuos para el gráfico de la curva ajustada
x_vals = np.linspace(1, 5, 100)
y_vals = modelo(x_vals, a1, a2, a3)

# Graficar los puntos observados y la curva ajustada
plt.scatter(x_data, y_data, color='red', label='Puntos observados')
plt.plot(x_vals, y_vals, color='blue', label=f'F(x) = {a1:.2f} + {a2:.2f}x + {a3:.2f}e^x')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Aproximación por Mínimos Cuadrados')
plt.legend()
plt.grid()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definir los puntos observados
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.2, 2.8, 3.6, 5.1, 7.3])

# Definir la función modelo F(x) = a1 + a2*x + a3*exp(x)
def modelo(x, a1, a2, a3):
    return a1 + a2*x + a3*np.exp(x)

# Definir la función modelo F(x) = a1 + a2*x
def lineales(x, b1, b2):
    return b1 + b2*x

# Ajuste de parámetros usando mínimos cuadrados no lineales
parametros, _ = curve_fit(modelo, x_data, y_data)

# Extracción de los coeficientes ajustados
a1, a2, a3 = parametros
print(f"Coeficientes ajustados: a1 ≈ {a1:.4f}, a2 ≈ {a2:.4f}, a3 ≈ {a3:.4f}")

# Calcular los valores ajustados
y_ajustado = modelo(x_data, a1, a2, a3)

# Calcular el coeficiente de determinación R^2
y_media = np.mean(y_data)
SST = np.sum((y_data - y_media) ** 2)  # Suma total de los cuadrados
SSR = np.sum((y_data - y_ajustado) ** 2)  # Suma de los cuadrados de los residuos
R2 = 1 - (SSR / SST)

print(f"Coeficiente de determinación R² ≈ {R2:.4f}")

# Extracción de los coeficientes ajustados
# Ajuste de parámetros usando mínimos cuadrados no lineales
parametros, _ = curve_fit(lineales, x_data, y_data)
b1, b2 = parametros
print(f"Coeficientes ajustados: b1 ≈ {b1:.4f}, b2 ≈ {b2:.4f}")

# Calcular los valores ajustados
y_ajustado_lineal = lineales(x_data, b1, b2)

# Calcular el coeficiente de determinación R^2
SSR = np.sum((y_data - y_ajustado_lineal) ** 2)  # Suma de los cuadrados de los residuos
R2 = 1 - (SSR / SST)

print(f"Coeficiente de determinación R² ≈ {R2:.4f}")



# Crear valores continuos para el gráfico de la curva ajustada
x_vals = np.linspace(1, 5, 100)
y_vals = modelo(x_vals, a1, a2, a3)
y_vals_lineal = lineales(x_vals, b1, b2)

# Graficar los puntos observados y la curva ajustada
plt.scatter(x_data, y_data, color='red', label='Puntos observados')
plt.plot(x_vals, y_vals, color='blue', label=f'F(x) = {a1:.2f} + {a2:.2f}x + {a3:.2f}e^x')
plt.plot(x_vals, y_vals_lineal, color='skyblue', label=f'F(x) = {b1:.2f} + {b2:.2f}x ')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Aproximación por Mínimos Cuadrados')
plt.legend()
plt.grid()
plt.show()
````
