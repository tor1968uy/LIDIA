#### **Métodos de Cuadratura: Método del Trapecio**

La integración numérica nos permite calcular aproximaciones de integrales definidas cuando no es posible resolverlas de forma analítica. Una de las técnicas más utilizadas es el método del trapecio, el cual se basa en aproximar el área bajo la curva mediante trapecios en lugar de rectángulos, proporcionando una mejor estimación en comparación con las sumas de Riemann.

##### **Ejemplo: Implementación del Método del Trapecio**

**Descripción del problema:**

Calcule el valor aproximado de la integral:
$$ I = \int_1^3 \ln(x)\, dx$$
utilizando el método del trapecio con $n=4$ subintervalos.

**Objetivo:**
- Implementar el método del trapecio en Python.
- Comparar el resultado con el valor exacto de la integral.

**Salidas:**
1. Aproximación de la integral usando el método del trapecio.
2. Comparación con el valor exacto.

```python
import numpy as np
import scipy.integrate as spi

# Definir la función a integrar
def f(x):
    return np.log(x)

# Implementación del método del trapecio
def metodo_trapecio(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)  # Generar puntos de la partición
    y = f(x)
    integral_aprox = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
    return integral_aprox

# Parámetros
a, b = 1, 3
n = 4

# Cálculo numérico
integral_aprox = metodo_trapecio(f, a, b, n)
integral_real, _ = spi.quad(f, a, b)  # Valor exacto usando SciPy

# Mostrar resultados
print(f"Aproximación por el método del trapecio: {integral_aprox:.6f}")
print(f"Valor exacto de la integral: {integral_real:.6f}")
print(f"Error absoluto: {abs(integral_real - integral_aprox):.6f}")
```

##### **Ejercicio 1: Aproximación de una Integral**

**Descripción del problema:**

Utilizando el método del trapecio, estime la siguiente integral:
$$ I = \int_0^1e^{-x^2} \, dx$$
para diferentes valores de $n$. Compare los resultados con el valor exacto obtenido numéricamente.

**Objetivo:**
- Implementar el método del trapecio en Python.
- Comparar el resultado con el valor exacto.
- Analizar cómo cambia la precisión al aumentar $n$.

**Entradas:**
- Diferentes valores de $n$ $(n=2, 4, 8, 16)$.

**Salidas:**
1. Valores aproximados de la integral para cada $n$.
2. Comparación con el valor exacto.


```python
import numpy as np
import scipy.integrate as spi

# Definir la función a integrar
def f(x):
    return np.exp(-x*x)        ## Funcion a Evaluar

# Implementación del método del trapecio
def metodo_trapecio(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)  # Generar puntos de la partición
                                  # crea un array con la cantidad de valores n+1 igualmente separados
                                  # entre lso valores a y b
    y = f(x)      # Evalua la funcion en lso puntos
    integral_aprox = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
    return integral_aprox

# Parámetros
a, b = 0, 1     # Valores de lso extremos de la integral definida

# Valor exacto de la integral (usando scipy)
integral_real, abserr  = spi.quad(f, a, b)

for n in [2, 4, 8, 16]:    # n tomando los valores indicados en la letra (2,4,8,16)
    # Cálculo numérico
    integral_aprox = metodo_trapecio(f, a, b, n)
    # Mostrar resultados
    print(f"===========================================================================================")
    print(f"Aproximación por el método del trapecio (n: {n}): {integral_aprox:.6f}")
    print(f"Valor exacto de la integral: {integral_real:.6f}, Error Absoluto de ScyPi {abserr:.2e}")
    print(f"Error absoluto: {abs(integral_real - integral_aprox):.4e}, Error relativo: {abs(integral_real - integral_aprox)/integral_real:.4e}")
```

#### **Métodos de Cuadratura: Método de Simpson**

El método de Simpson mejora la precisión de la integración numérica al utilizar parábolas en lugar de segmentos de recta para aproximar la función. Esta técnica se basa en la interpolación de tres puntos consecutivos con un polinomio cuadrático, lo que permite capturar mejor la curvatura de la función y reducir el error de aproximación.

##### **Ejemplo: Implementación del Método de Simpson**

**Descripción del problema:**
Vamos a calcular la integral:
$$ I = \int_0^2e^{-x^2} \, dx$$
utilizando el método de Simpson con $n=2$.

**Objetivo:**
- Implementar el método de Simpson en Python.
- Comparar el resultado con el valor exacto.



```python
import numpy as np
import scipy.integrate as spi

def simpson(f, a, b, n):
    # Implementa el método de Simpson para aproximar la integral definida de f en [a, b] con n pares de subintervalos.
    # Parámetros: f (función a integrar),  a (límite inferior de integración), b (límite superior de integración), n (número de intervalos (debe ser par))
    # Retorna: float : Aproximación de la integral mediante el metodo de Simpson

    if n % 2 == 1:
        raise ValueError("n debe ser par para el método de Simpson")

    h = (b - a) / n  # Tamaño del subintervalo
    x = np.linspace(a, b, n+1)  # Generar puntos de la partición
                                # crea un array con la cantidad de valores n+1 igualmente separados
                                # entre los valores a y b
    y = f(x)                    # Evalua la función en los puntos
    integral = (h/3) * (y[0] + 4 * sum(y[1:n:2]) + 2 * sum(y[2:n-1:2]) + y[n]) # sum(y[1:n:2]) es el valor de los intervalos impares
                                                                                # sum(y[2:n-1:2]) es el valor de los intervalos pares

    return integral

# Definimos la función a integrar
def funcion(x):
    return np.exp(-x**2)  # Funcion a Integrar

# Intervalo de integración
a, b = 0, 2  # limite inferior y superior de integracion

# Número de pares de subintervalos
n = 2

# Cálculo de la integral usando Simpson
resultado = simpson(funcion, a, b, n)

# Valor exacto de la integral (usando scipy)
I_exacto, abserr = spi.quad(funcion, a, b)

# Mostramos los resultados
print(f"Aproximación con Simpson (n={n}): {resultado:.6f}")
print(f"Valor exacto de la integral: {I_exacto:.6f}, Error absoluto del calculo de ScyPi {abserr:.2e}")
print(f"Error absoluto: {abs(resultado - I_exacto):.4e}, Error relativo: {abs(resultado - I_exacto)/I_exacto:.4e}")
```

Los métodos adaptativos de integración numérica ajustan dinámicamente el número de puntos de evaluación dependiendo del comportamiento de la función. A diferencia de los métodos clásicos como el del trapecio o de Simpson, que usan particiones uniformes, los métodos adaptativos como Gauss-Kronrod asignan más puntos a regiones donde la función cambia rápidamente, logrando así mayor precisión con menos evaluaciones totales.

#### **Introducción a la cuadratura de Gauss**

La cuadratura de Gauss permite aproximar una integral definida $\int_a^b f(x) dx$ usando una combinación ponderada de evaluaciones de $f$ en puntos estratégicos:

$$\int_a^b f(x) dx \approx \sum w_i f(x_i)$$

Los nodos $(x_i)$ y pesos $(w_i)$ son elegidos para maximizar la precisión en polinomios de hasta cierto grado. Esto la convierte en una técnica muy eficiente para funciones suaves.

##### **Ejemplo: Cuadratura de Gauss (2 puntos)**

**Descripción del problema:**

Aproxime la integral
$$ I=\int_0^2e^{-x^2}dx$$
utilizando la **cuadratura de Gauss con 2 puntos**, y compare el resultado con el valor exacto aproximado usando `scipy.integrate.quad`.

**Objetivos:**
- Implementar la fórmula de la cuadratura de Gauss con 2 puntos en el intervalo $[a,b]$.
- Aplicar un cambio de variable para adaptar la fórmula al intervalo $[0,2]$.
- Comparar la aproximación con el valor exacto numérico.
- Calcular y mostrar el error absoluto.

```python
import numpy as np

def gauss_2pts(f, a, b):
    x0 = -(1/np.sqrt(3))
    x1 = -x0
    w0 = w1 = 1
    # Cambio de variable
    t0 = (b-a)/2 * x0 + (a+b)/2
    t1 = (b-a)/2 * x1 + (a+b)/2
    return (b-a)/2 * (w0*f(t0) + w1*f(t1))

# Ejemplo de uso
f = lambda x: np.exp(-x**2)
aprox = gauss_2pts(f, 0, 2)
print("Aproximación con Gauss 2 pts:", round(aprox, 6))
```

#### **Métodos adaptativos: Método de Gauss-Kronrod**

El método de Gauss-Kronrod amplía la cuadratura de Gauss agregando nodos intermedios para obtener una estimación más precisa sin duplicar las evaluaciones. Además, permite estimar el error comparando la aproximación de Gauss con la de Gauss-Kronrod.

##### **Ejemplo: Gauss-Kronrod (G7, K15)**

**Descripción del problema:**

Aproxime la integral
$$I = \int_1^3 \cos(x) dx$$
utilizando el método de Gauss-Kronrod, con los nodos y pesos de las reglas $G_7$ y $K_{15}$.

**Objetivos:**
- Reescribir la integral al intervalo estándar $[-1,1]$ usando un cambio de variable adecuado.
- Evaluar la integral utilizando simultáneamente la cuadratura de Gauss (7 puntos) y la de Gauss-Kronrod (15 puntos).
- Comparar ambas aproximaciones y estimar el error a partir de la diferencia entre los métodos.
- Discutir la eficiencia del método de Gauss-Kronrod para obtener alta precisión con una cantidad controlada de evaluaciones.

```python
from math import cos

def gk15(f):
    # Nodos y pesos de Gauss-Kronrod G7 y K15
    xk = [
        0.9914553711208126, 0.9491079123427585, 0.8648644233597691,
        0.7415311855993945, 0.5860872354676911, 0.4058451513773972,
        0.2077849550078985, 0.0
    ]
    wk = [
        0.0229353220105292, 0.0630920926299786, 0.1047900103222502,
        0.1406532597155259, 0.1690047266392679, 0.1903505780647854,
        0.2044329400752989, 0.2094821410847278
    ]
    # Duplicación para reflejar en [-1,1]
    res_k = 0.0
    res_g = 0.0
    for i in range(7):
        res_k += wk[i]*f(xk[i]) + wk[i]*f(-xk[i])
    res_k += wk[7]*f(0.0)

    # Pesos de Gauss G7 (subset de K15)
    xg = [0.9491079123427585, 0.7415311855993945, 0.4058451513773972, 0.0]
    wg = [0.1294849661688697, 0.2797053914892766, 0.3818300505051189, 0.4179591836734694]
    for i in range(3):
        res_g += wg[i]*f(xg[i]) + wg[i]*f(-xg[i])
    res_g += wg[3]*f(0.0)

    return res_g, res_k, abs(res_k - res_g)

# Cambio de variable para evaluar en [1,3]
def f_transformada(x):
    return cos((3-1)/2*x + (3+1)/2)

G, K, error = gk15(f_transformada)
print("Gauss G7:", round(G, 10))
print("Gauss-Kronrod K15:", round(K, 10))
print("Error estimado:", error)
```

##### **Ejercicio 5: Estimación de $\displaystyle \int_0^5 \frac{\sin(x)}{x+1} dx$ con Gauss-Kronrod**

**Descripción del problema:**

Utilizaremos el método de Gauss-Kronrod con $n=7$ para aproximar la integral
$$I = \int_0^5\frac{\sin(x)}{x+1} dx$$
La función presenta una oscilación moderada en el intervalo, lo que hace interesante evaluar el desempeño del método adaptativo en este caso.

**Objetivos:**
- Aplicar un cambio de variable para transformar la integral al intervalo $[-1,1]$.
- Aproximar la integral usando la regla de Gauss-Kronrod $K_{15}$ y comparar con `scipy.integrate.quad`.
- Calcular el error absoluto.
- Extra: Graficar el error absoluto comparando diferentes métodos si el tiempo lo permite (por ejemplo, Gauss 2 puntos, Simpson, etc.).

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

def gk15(f_transformed, a_original, b_original):
    # Nodos y pesos de Gauss-Kronrod K15
    xk = [
        0.9914553711208126, 0.9491079123427585, 0.8648644233597691,
        0.7415311855993945, 0.5860872354676911, 0.4058451513773972,
        0.2077849550078985, 0.0
    ]
    wk = [
        0.0229353220105292, 0.0630920926299786, 0.1047900103222502,
        0.1406532597155259, 0.1690047266392679, 0.1903505780647854,
        0.2044329400752989, 0.2094821410847278
    ]

    # Nodos y pesos de Gauss G7 (subset de K15)
    xg = [0.9491079123427585, 0.7415311855993945, 0.4058451513773972, 0.0]
    wg = [0.1294849661688697, 0.2797053914892766, 0.3818300505051189, 0.4179591836734694]

    # Factor de escalado debido al cambio de variable dx = ((b-a)/2)dt
    scale_factor = (b_original - a_original) / 2

    res_k = 0.0
    for i in range(7): # Suma para los nodos no centrales (simétricos)
        res_k += wk[i] * (f_transformed(xk[i]) + f_transformed(-xk[i]))
    res_k += wk[7] * f_transformed(0.0) # Suma para el nodo central
    res_k *= scale_factor # Aplicar el factor de escalado

    res_g = 0.0
    for i in range(3): # Suma para los nodos no centrales (simétricos)
        res_g += wg[i] * (f_transformed(xg[i]) + f_transformed(-xg[i]))
    res_g += wg[3] * f_transformed(0.0) # Suma para el nodo central
    res_g *= scale_factor # Aplicar el factor de escalado

    return res_g, res_k, abs(res_k - res_g)

def gauss_2pts(f, a, b):
    x0 = -1 / np.sqrt(3)
    x1 = -x0
    w0 = 1
    w1 = 1
    t0 = (b-a)/2 * x0 + (a+b)/2
    t1 = (b-a)/2 * x1 + (a+b)/2
    return (b-a)/2 * (w0*f(t0) + w1*f(t1))

def simpson(f, a, b, n=1000):
    if n % 2 != 0:
        n += 1 # n debe ser par para Simpson
    h = (b - a) / n
    sum_val = f(a) + f(b)
    for i in range(1, n, 2):
        sum_val += 4 * f(a + i * h)
    for i in range(2, n - 1, 2):
        sum_val += 2 * f(a + i * h)
    return sum_val * h / 3

# Función original a integrar
def f_original(x):
    return np.sin(x) / (x + 1)

# Límites de integración originales
a_original = 0
b_original = 5

# 1. Aplicar un cambio de variable para transformar la integral a [-1, 1]
# La transformación es: x = (b-a)/2 * t + (a+b)/2
# Donde t es la variable en [-1, 1]
# La nueva función a integrar (f_transformada(t)) será f_original(x(t))
def f_transformada(t):
    x = (b_original - a_original) / 2 * t + (a_original + b_original) / 2
    return f_original(x)

# Aproximación con Gauss-Kronrod
G_approx, K_approx, estimated_error_gk = gk15(f_transformada, a_original, b_original)

print(f"Aproximación con Gauss G7:            {G_approx:.10f}")
print(f"Aproximación con Gauss-Kronrod K15:   {K_approx:.10f}")
print(f"Error estimado (abs(K15 - G7)):       {estimated_error_gk:.10f}\n")

# Valor exacto aproximado con scipy.integrate.quad
valor_exacto_scipy, error_scipy = integrate.quad(f_original, a_original, b_original)
print(f"Valor exacto aproximado (scipy.integrate.quad): {valor_exacto_scipy:.10f}")
print(f"Error estimado por scipy.integrate.quad:        {error_scipy:.10e}\n")

# Calcular el error absoluto de Gauss-Kronrod K15
error_absoluto_K15 = abs(valor_exacto_scipy - K_approx)
print(f"Error absoluto de K15 respecto a SciPy:         {error_absoluto_K15:.10f}\n")


# Aproximación con Gauss 2 puntos
aprox_gauss_2pts = gauss_2pts(f_original, a_original, b_original)
error_absoluto_gauss_2pts = abs(valor_exacto_scipy - aprox_gauss_2pts)
print(f"Aproximación con Gauss 2 puntos:      {aprox_gauss_2pts:.10f}")
print(f"Error absoluto de Gauss 2 puntos:     {error_absoluto_gauss_2pts:.10f}\n")

# Aproximación con Regla de Simpson (con N grande para buena precisión)
aprox_simpson = simpson(f_original, a_original, b_original, n=1000)
error_absoluto_simpson = abs(valor_exacto_scipy - aprox_simpson)
print(f"Aproximación con Simpson (N=1000):    {aprox_simpson:.10f}")
print(f"Error absoluto de Simpson (N=1000):   {error_absoluto_simpson:.10f}\n")

# Gráfica del error absoluto
metodos = ['Gauss 2 pts', 'Simpson (N=1000)', 'Gauss-Kronrod K15']
errores = [error_absoluto_gauss_2pts, error_absoluto_simpson, error_absoluto_K15]

plt.figure(figsize=(10, 6))
plt.bar(metodos, errores, color=['skyblue', 'lightgreen', 'salmon'])
plt.yscale('log') # Usamos escala logarítmica para apreciar las diferencias de orden de magnitud
plt.title('Error Absoluto de Diferentes Métodos de Cuadratura')
plt.xlabel('Método de Integración')
plt.ylabel('Error Absoluto (Escala Logarítmica)')
plt.grid(True, which="both", ls="--", c='0.7')
plt.show()
```
